```python
import pandas as pd
```

```python
df=pd.read_csv('startup_funding.csv')
```

```python
df.info()
```

```python
#  removing remarks because too much missing values and making it permanent change

df.drop(columns=['Remarks'],inplace =True)
```

```python
# making Sr No the index of this Dataframe

df.set_index('Sr No', inplace = True)
df
```

```python
#  changing columns name

df.rename(columns={
    'Date dd/mm/yyyy':'date',
    'Startup Name':'startup',
    'Industry Vertical':'vertical',
    'SubVertical':'subvertical',
    'City  Location':'city',
    'Investors Name':'investors',
    'InvestmentnType':'round',
    'Amount in USD':'amount' 
},inplace = True)
df
```

```python
df.info()

# we can see there are many missing values in amount.(because money raised by the investors are undisclosed)
```

```python
df['amount'] = df['amount'].fillna('0')

df.info()
```

```python
#  converting into integer

# df['amount'].astype('int')     #cant do this because the amount is in string with commas


# Remove commas
df['amount'] = df['amount'].str.replace(',', '')

# Convert to numeric, converting non-numeric strings to NaN
df['amount'] = pd.to_numeric(df['amount'], errors='coerce')

# Optional: Handle NaN values
# For example, replace NaN values with 0
df['amount'] = df['amount'].fillna(0)

# Convert to int
df['amount'] = df['amount'].astype(int)

df.info()
```

```python
df['date']
```

```python
#  converting date column into date-time object

       
df['date'] = pd.to_datetime(df['date'], format='%d/%m/%Y', errors= 'coerce') 
df['date']
# since our data have date in the format of d/m/y so we have to specify this.
```

```python
df['date'].dt.day

#  extracting day
```

```python
df.info()
```

```python
#  removing the rows where we have missing values.

df.dropna(subset=['date','startup','vertical','subvertical','city','investors','round','amount'],inplace=True)
```
